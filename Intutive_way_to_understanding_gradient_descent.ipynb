# Concept of  Gradient


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Define the gradient function of f(x, y)
def gradient_f(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

# Generate data for plotting
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Calculate gradient at point (1, 2)
gx, gy = gradient_f(1, 2)

# Plot the function surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.set_title('Function f(X, Y) = X^2 + Y^2')

# Plot the point (1, 2, f(1, 2)) on the surface
ax.scatter([1], [2], [f(1, 2)], color='red', s=100, label='Point (1, 2, f(1, 2))')

# Plot the gradient vector (2, 4) at point (1, 2)
ax.quiver(1, 2, f(1, 2), gx, gy, 0, color='green', label='Gradient (2, 4)', length=1, normalize=True)

# Adjust plot limits and show legend
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([0, 18])
ax.legend()

plt.show()
```

Gradient Descent Visualization

```python
import numpy as np
import matplotlib.pyplot as plt

# Function J(theta)
def J(theta):
    return theta ** 2

# Gradient of J(theta)
def grad_J(theta):
    return 2 * theta

# Gradient descent algorithm
def gradient_descent(theta0, alpha, iterations):
    theta = theta0
    theta_history = [theta]

    for _ in range(iterations):
        gradient = grad_J(theta)
        theta = theta - alpha * gradient
        theta_history.append(theta)

    return np.array(theta_history)

# Parameters
theta0 = 4.0  # Initial guess
alpha = 0.1   # Learning rate
iterations = 20

# Perform gradient descent
theta_history = gradient_descent(theta0, alpha, iterations)

# Plotting
theta_vals = np.linspace(-5, 5, 100)
J_vals = J(theta_vals)

plt.figure(figsize=(10, 6))
plt.plot(theta_vals, J_vals, label='J(theta) = theta^2')
plt.scatter(theta_history, J(np.array(theta_history)), color='red', label='Gradient Descent Path')
plt.title('Gradient Descent Visualization')
plt.xlabel('theta')
plt.ylabel('J(theta)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function J(theta)
def J(theta):
    return theta[0]**2 + theta[1]**2

# Gradient of J(theta)
def grad_J(theta):
    return np.array([2 * theta[0], 2 * theta[1]])

# Gradient descent algorithm
def gradient_descent(theta0, alpha, iterations):
    theta = theta0
    theta_history = [theta]

    for _ in range(iterations):
        gradient = grad_J(theta)
        theta = theta - alpha * gradient
        theta_history.append(theta)

    return np.array(theta_history)

# Parameters
theta0 = np.array([4.0, 4.0])  # Initial guess (starting point)
alpha = 0.1   # Learning rate
iterations = 20

# Perform gradient descent
theta_history = gradient_descent(theta0, alpha, iterations)

# Plotting the 3D surface of J(theta)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)
theta0_mesh, theta1_mesh = np.meshgrid(theta0_vals, theta1_vals)
J_vals = theta0_mesh**2 + theta1_mesh**2

ax.plot_surface(theta0_mesh, theta1_mesh, J_vals, cmap='viridis', alpha=0.8)
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J(theta)')
ax.set_title('Gradient Descent in 3D')

# Plotting the gradient descent path
theta_history = np.array(theta_history)
ax.plot(theta_history[:, 0], theta_history[:, 1], J(theta_history.T), marker='o', color='r', label='Gradient Descent Path')
ax.legend()

plt.show()
```
